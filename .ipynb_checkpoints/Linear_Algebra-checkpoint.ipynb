一个向量中成分的个数就是这个向量的维数。

列向量有助于像两坐标变换，空间映射等计算过程，因此更受欢迎。


```python
a = np.array([1,2,3,4])
print(a)
```

```python
a = np.array([1,2,3,4])
print(a.transpose()) #didn't work like this
```

```python
a = np.array([1,2,3,4])
a_t = a[:,np.newaxis]
print(a_t) # there are easy ways to do it
```

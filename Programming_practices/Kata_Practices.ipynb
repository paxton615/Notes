```python
# my code, passed the test
def series_sum(n):    
    import numpy as np
    num = n
    counting = 0
    if num == 0:
        return (str("{:.2f}".format(num)))
    else:
        for i in range(1,n,1):
            counting += 1/(4*i-(i-1))
        return (str("{:.2f}".format(np.round(1+counting , 2))))
```

```python
series_sum(5)
```

```python
def high_and_low(numbers):
    numlist = numbers.split(" ")
    i = 0
    highest = int(numlist[0])
    lowest = int(numlist[0])
    while i < len(numlist):
        numlist[i] = int(numlist[i])
        if numlist[i] > highest:
            highest = numlist[i]
        if numlist[i] < lowest:
            lowest = numlist[i]
        i += 1
    highest = str(highest)
    lowest = str(lowest)
    return  highest+" "+lowest
```

```python
def ticket(people):
    order = 0
    erwu = 0
    wushi = 0
    yibai = 0
    while order < len(people):
        if people[order] == 25:
            erwu += 1

        if people[order] == 50:
            erwu -= 1
            wushi += 1

        if people[order] == 100:
            erwu -= 1
            wushi -= 1
            yibai += 1

        order += 1

    if all(i>=0 for i in [erwu, wushi, yibai]) == True:  # pretty useful function all( )
        return('YES')
    else:
        return('NO')

    # print(erwu)
    # print(wushi)
    # print(yibai)
    
```

```python
def spin_words(sentence):    
    def reverse_words(ci):
        # Your code goes here
        str_rev = list(ci)
        str_rev.reverse()
        n=0
        letters=''
        while n < len(str_rev):
            letters += str_rev[n]
            n +=1
        return (letters)

    new_sentence = []
    for i in sentence.split():
    #     print(i)
        if len(i) < 5:
            new_sentence.append(i)
        else:
            new_sentence.append(reverse_words(i))
#     return(new_sentence) #return

    new_sen = ''
    tick = 0
    while tick < len(new_sentence):
        new_sen += new_sentence[tick]+' '
        tick += 1
    return(new_sen.strip())
```

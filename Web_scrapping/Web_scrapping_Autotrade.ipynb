```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = 'https://www.autotrader.co.uk/car-search?postcode=GU216DR&make=AUDI&model=Q2&homeDeliveryAdverts=include&fuel-type=Petrol&year-to=2019&onesearchad=Used&onesearchad=Nearly%20New&onesearchad=New&advertising-location=at_cars&include-delivery-option=on&page=1'
r = requests.get(url)
```

```python
r.text[:1000]
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.find_all('li', class_='search-page__result')
```

```python
for tag in soup.find_all('h2'):
    print( tag.text )
```

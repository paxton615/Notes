```python
from flask import Flask, render_template, request
import json

app = Flask(__name__)


@app.route('/')        # 重要，指定了url的访问路径
def hello_world():     # 定义这个url中的内容
    return 'Hello World!'

@app.route('/hello')   # 更换路径
def hello_world2():
    data = 'hello data'
    return render_template('hello.html', data = data)

# 路由route的本质就是个装饰器
@app.route('/user/<username>', methods=['get', 'post'])
def get_user(username):
    return 'hello {}'.format(username)



@app.route('/data', methods=['get', 'post'])
def test_data():
    # print(request.args)
    # print(request.args.get('a'), request.args.get('b'))
    # print(request.headers)
    # print(request.headers.get('User-Agent'))
    # print(request.data)
    # import json
    # print(json.loads(request.data))
    # print(request.cookies)
    # print(request.cookies.get('token'))
    # print(request.form)
    # print(request.form.get('username'), request.form.get('password'))
    return 'success'

@app.route('/user_template')
def use_template():
    datas = [(1, 'name1'),(2, 'name2'), (3, 'name3')]
    title = '学生信息'
    return render_template('user_template.html', datas = datas, title = title)

def read_pvuv_data():
    '''
    read pvuv data
    :return: list, ele:(pdate, pv, uv)
    '''
    data = []
    with open('./data/pv_uv.txt') as fin:
        for line in fin:
            pdate, pv, uv = line.split('-')
            data.append((pdate, pv, uv))
    return (data)

@app.route('/pvuv')
def pvuv():
    # read file
    data = read_pvuv_data()
    # return html
    return render_template('pvuv.html', data = data)

@app.route('/getjson')
def getjson():
    # read file
    data = read_pvuv_data()
    # return html
    return json.dumps(data)
# 虽然我们开发的是一个后端的api，但是在前端也是可以访问的

if __name__ == '__main__':
    app.run()
```

```python
import pymysql

# 用pymysql实现pycharm与mysql的connection
def get_conn():
    return pymysql.connect(   # 这里开头直接是return
        host = '127.0.0.1',
        user = 'root',
        password = 'pcx19570403',
        database = 'python_mysql',
        charset = 'utf8'
    ) # 先测试函数内容，RUN后不报错,说明成功,再用函数封装，以便之后使用

# 用一个函数来提取sql请求，抓取数据
def query_data(sql):
    conn = get_conn()  # 建立连接
    try:
        cursor = conn.cursor(pymysql.cursors.DictCursor) #在建好的链接中,创建pymysql光标，并返回字典形式
        cursor.execute(sql)  # 光标执行sql语句
        return cursor.fetchall() # 提取所有内容
    finally:
        conn.close() # 为防止资源占用,使用try, finally使用后关闭

# 用一个函数来插入，更新mysql数据库
def insert_or_update_data(sql): # 插入和更新表格
    conn = get_conn() # 建立连接
    try:
        cursor = conn.cursor() #建立光标
        cursor.execute(sql) #光标执行sql语句
        conn.commit()  # 确认commit
    finally:
        conn.close()

# 统一执行以上出现过的函数
if __name__ == '__main__':
    # 插入，更新一条sql数据
    sql = 'insert user (name, sex, age, email) values("daming", "male", 23 , "daming@gmail.com")' # 内容
    insert_or_update_data(sql) # 执行函数
    
    # 提取mysql数据并展示
    sql = 'select * from user'
    datas = query_data(sql)
    import pprint
    pprint.pprint(datas)
```

```python
# my code, passed the test
def series_sum(n):    
    import numpy as np
    num = n
    counting = 0
    if num == 0:
        return (str("{:.2f}".format(num)))
    else:
        for i in range(1,n,1):
            counting += 1/(4*i-(i-1))
        return (str("{:.2f}".format(np.round(1+counting , 2))))
```

```python
series_sum(5)
```

```python
def find_next_square(sq):
    # Return the next square if sq is a square, -1 otherwise
    import numpy as np
    if sq**0.5%1 == 0: 
        return (np.square(sq**0.5+1))
    else:
        return -1
```

```python
def find_next_square(sq):
    root = sq ** 0.5
    if root.is_integer():  # 判断是否是整数的方法！！！！！！
        return (root + 1)**2
    return -1
```

```python
def tower_builder(n):
    # build here
    tower=[]
    for i,j in zip(range(n)[::-1],range(n)):
        tower.append((' '*i)+('*'*int(j*2+1))+(' '*i))
    return tower
```

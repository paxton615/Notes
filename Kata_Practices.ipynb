```python
# my code, passed the test
def series_sum(n):    
    import numpy as np
    num = n
    counting = 0
    if num == 0:
        return (str("{:.2f}".format(num)))
    else:
        for i in range(1,n,1):
            counting += 1/(4*i-(i-1))
        return (str("{:.2f}".format(np.round(1+counting , 2))))
```

```python
series_sum(5)
```

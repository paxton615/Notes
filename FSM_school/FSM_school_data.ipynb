```python
import numpy as np
import pandas as pd


import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
year = [2017, 2018,2019, 2020, 2021 ,2022, 2023 ,2024]
student_num = [222, 186, 170, 159, 117, 152, 178, 216]
net_income = [3250000, 2942000, 2885000 ,2591359, 1876316, 2616089, 3350775, 3966441]
total_cost = [3655000, 3280000, 3201000, 2514047, 2383116, 3082151, 3356000, 3591550]
```

```python
df = pd.DataFrame(list(zip(net_income, total_cost)), index=[2017, 2018,2019, 2020, 2021 ,2022, 2023 ,2024])
```

```python
df.columns = ['net_income','total_cost']
```

```python
df['profit'] = df.net_income - total_cost
df['student_num'] = student_num
```

```python
df
```

```python
df['net_income_per_stu'] = df.net_income / df.student_num
df['cost_per_student'] = df.total_cost / df.student_num
df['profit_per_student'] = df.profit / df.student_num
```

```python
df
```

```python
df.describe()
```

```python
from sklearn.linear_model import LinearRegression

mod = LinearRegression()
mod.fit(df[['total_cost']], df[['net_income']])

print(mod.intercept_)
print(mod.coef_)
```

```python
from matplotlib.pyplot import figure
figure(figsize=(8, 6), dpi=80)
sns.regplot(df[['total_cost']], df[['net_income']]);
```

### Perdiction: Student_num vs net_income

```python
from matplotlib.pyplot import figure
figure(figsize=(8, 6), dpi=80)
sns.regplot(df[['student_num']], df[['net_income']]);
```

```python
from sklearn.linear_model import LinearRegression

mod = LinearRegression()
mod.fit(df[['student_num']], df[['net_income']])

print(mod.intercept_)
print(mod.coef_)
print(mod.predict([[300]]))
```

```python
Price = {"Date":[],'USD(AM)':[],'USD (PM)':[],'GBP (AM)':[],
         'GBP (PM)':[],'EURO (AM)':[],'EURO (PM)':[]}

for item in json_['dataset']["data"]:
    
    try:
        Price['Date'].append(item[0])
    except:
        Price['Date'].append('None')
    
    
    try:
        Price['USD(AM)'].append(item[1])
    except:
        Price['USD(AM)'].append('None')

    try:
        Price['USD (PM)'].append(item[2])
    except:
        Price['USD (PM)'].append('None')
        
    try:
        Price['GBP (AM)'].append(item[3])
    except:
        Price['GBP (AM)'].append('None')
        
    try:
        Price['GBP (PM)'].append(item[4])
    except:
        Price['GBP (PM)'].append('None')
        
    try:
        Price['EURO (AM)'].append(item[5])
    except:
        Price['EURO (AM)'].append('None')
    
    try:
        Price['EURO (PM)'].append(item[6])
    except:
        Price['EURO (PM)'].append('None')
        
Price = pd.DataFrame(Price)
Price
```

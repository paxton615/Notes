### 2. Join the bikeshare dataframe and the weather dataframe

```python
bikeshare["weather_2"]=weather["weather_2"]
bikeshare["weather_3"]=weather["weather_3"]
bikeshare["weather_4"]=weather["weather_4"]
```

```python
bikeshare['year']= bikeshare.dteday.apply(lambda x: float((str(x).split('-'))[0]))
bikeshare['month']=bikeshare.dteday.apply(lambda x: float((str(x).split('-'))[1]))
bikeshare.head(5)
```

```python
bikeshare = bikeshare[[ 'season', 'mnth', 'hr', 
        'weathersit', 'temp', 'atemp', 'hum', 'windspeed',
        'weather_2', 'weather_3', 'weather_4'
       ]]
```

### 3. Choose the weatherrelated features as predictors and the season as response

```python
X=bikeshare[[
         'temp', 'atemp', 'hum', 'windspeed',
        'weather_2', 'weather_3', 'weather_4']]
y=bikeshare["season"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.5, random_state=1)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```

## AR(p) process:

General form:

$$
\begin{eqnarray*}
y_t&=&\phi_0+\phi_1 y_{t-1}+\ldots+\phi_p y_{t-p}+\epsilon_t\\
\phi_0&=&(1-(\phi_1+\ldots+\phi_p))\mu
\end{eqnarray*}
$$

#### An AR(1) example

For an AR(1)-model, this equation would look like

$$
\begin{eqnarray*}
y_t &=& \phi_0 + \phi_1  y_{t-1} +\ \epsilon_t
\end{eqnarray*}
$$

with 

$$\phi_0=\left(1-\phi_1\right)\mu$$

Let's say, we have a single starting value for a time series, we know the mean and the coefficient $\phi_1$. Then we can predict all the other points in the time series up to the residual term.

Consider for example the values

$$
\begin{eqnarray*}
\mu &=& 1\\
\phi_1 &=& -0.8\\
y_0 &=& 0
\end{eqnarray*}
$$

That means 

$$
\phi_0 = (1-(-0.8))\cdot 1 = 1.8
$$

Then for $y_1$ we would predict

$$
y_1 = 1.8 + (-0.8)\cdot 0 + \epsilon_1
$$

For $\epsilon_t$, we will always make the prediction
$\epsilon_t=0$ as it is completely random, so we do not know better than to predict its mean, which is zero.
Therefore we predict

$$
y_1 = 1.8
$$

In the same way, we would predict $y_2$, being ignorant about the noise term, based on $y_1$:

$$
y_2 = 1.8 + (-0.8)\cdot 1.8 = 0.36
$$

We could go on to predict all the other points. All what we needed to do so were the coefficients and one starting value for the time series.

### Fit an AR(3) process with

$\mu=104$, $\phi_1=0.4$, $\phi_2=0.25$, $\phi_3=0.1$

$y_{n-3}=105$, $y_{n-2}=102$, $y_{n-1}=103$, $y_{n}=99$

What are $y_{n+1}$ and $y_{n+2}$?

#### Bonus: Write a function which takes $\mu$, $\phi$, the known $y_t$ and a forecast length as arguments and returns a list of the time series including the predicted values.

```python
mu = 104
phi = np.array([0.4, 0.25, 0.1])
y = np.array([99., 103., 102., 105.])
n = 10


def ar_forecast(mu, phi, y, n):
    phi_0 = (1-np.sum(phi))*mu
    i = 0
    while i < n:
        y_next = phi_0 + phi.dot(y[:len(phi)])
        y = np.insert(y, 0, y_next)
        i += 1
    return y


print(ar_forecast(mu, phi, y, n))
```

```python
plt.plot(ar_forecast(mu, phi, y, 20)[-1::-1])
plt.show()
```

## MA(q) process:
$$y_t=\mu+\epsilon_t+\theta_1\epsilon_{t-1}+\ldots+\theta_q\epsilon_{t-q}$$

#### MA(1) example

Let's say we have 

$$
\begin{eqnarray*}
\mu &=& 10 \\
\theta_1 &=& 0.5 \\
y_0 &=& 5 \\
\epsilon_0 &=& 3
\end{eqnarray*}
$$

Then we would predict

$$
\begin{eqnarray*}
y_1 &=& \mu + \theta_1\cdot \epsilon_{0}
= 10+0.5\cdot 3 = 11.5\\
\epsilon_1 &=& 0 
\end{eqnarray*}
$$

predicting again the mean (i.e. zero) for the random noise term $\epsilon_1$.

Then 

$$
\begin{eqnarray*}
y_2 &=& \mu + \theta_1\cdot \epsilon_{1}
= 10+0.5\cdot 0 = 10\\
y_3 &=& 10 
\end{eqnarray*}
$$

and so forth.

### Fit an MA(2) process with

$\mu=45$,
$\theta_1=0.3$,
$\theta_2=-0.15$,


$y_n=42.7$,
$y_{n-1}=39.8$,


$\epsilon_n=1.5$,
$\epsilon_{n-1}=-4.3$.

What are $y_{n+1}$ and $y_{n+2}$?

Hint: Remember that $\epsilon_t$ are supposed to result as noise from a normal distribution with mean zero.

#### Bonus: Write a function which takes $\mu$, $\theta$, the known $y_t$, the known $\epsilon_t$ and a forecast length as arguments and returns a list of the time series including the predicted values.

```python
mu = 45
theta = np.array([0.3, -0.15])
y = np.array([42.7, 39.8])
epsilon = np.array([1.5, -4.3])
n = 5


def ma_forecast(mu, theta, y, epsilon, n):
    i = 0
    while i < n:
        y_next = mu + theta.dot(epsilon[:len(theta)])
        epsilon = np.insert(epsilon, 0, 0)
        y = np.insert(y, 0, y_next)
        i += 1
        print(epsilon)
    return y


print(ma_forecast(mu, theta, y, epsilon, n))
```

```python
plt.plot(ma_forecast(mu, theta, y, epsilon, n)[-1::-1])
plt.show()
```

## ARMA(p,q) process:
$$y_t=\phi_0+\phi_1 y_{t-1}+\ldots+\phi_p y_{t-p}+\epsilon_t+\theta_1\epsilon_{t-1}+\ldots+\theta_q\epsilon_{t-q}$$

### Fit an ARMA(1,2) process with

$\mu=103$,
$\phi_1=0.2$,
$\theta_1=0.4$,
$\theta_2=-0.25$,
$y_n=120.1$,
$y_{n-1}=118.3$,
$\epsilon_n=-2.3$,
$\epsilon_{n-1}=2.6$.

What are $y_{n+1}$ and $y_{n+2}$?

#### Bonus: Write a function which takes $\mu$, $\phi$, $\theta$, the known $y_t$, the known $\epsilon_t$ and a forecast length as arguments and returns a list of the time series including the predicted values.

```python
mu = 103
phi = np.array([0.2])
theta = np.array([0.4, -0.25])
y = np.array([120.1, 118.3])
epsilon = np.array([-2.3, 2.6])
phi_0 = (1-np.sum(phi))*mu
#phi_0 = mu
n = 10


def arma_forecast(phi_0, phi, theta, y, epsilon, n):
    i = 0
    while i < n:
        y_next = phi_0 + phi.dot(y[:len(phi)]) + theta.dot(
            epsilon[:len(theta)])
        epsilon = np.insert(epsilon, 0, 0)
        y = np.insert(y, 0, y_next)
        i += 1
    return y


print(arma_forecast(phi_0, phi, theta, y, epsilon, n))
```

```python
plt.plot(arma_forecast(phi_0, phi, theta, y, epsilon, n)[-1::-1])
plt.show()
```

### 4. Load the CRSP daily returns. Fit an ARMA model to the GE column and tune the parameters.

#### Bonus: Fit also the other columns.

#### Obtain a date index by combining the year, month and day data.

```python
df = pd.read_csv('datasets/CRSPday.csv')
df['date'] = df.year.astype(
    'str')+'/'+df.month.astype('str')+'/'+df.day.astype('str')
df['date'] = pd.to_datetime(df.date)
df.set_index(df.date, inplace=True)
df.head()
```

```python
df.shape
```

#### Plot the timeseries and the (partial) autocorrelations

```python
fig, ax = plt.subplots(ncols=3, figsize=(20, 6))
df['ge'].plot(ax=ax[0], title='GE time series')
plot_acf(df['ge'], lags=10, alpha=0.05, ax=ax[1])
plot_pacf(df['ge'], lags=10, alpha=0.05, ax=ax[2])
plt.show()
```

#### Fit the ARMA model

```python
auto_select = sm.tsa.arma_order_select_ic(df['ge'].values,
                                          max_ar=5, max_ma=2, ic=['aic'],
                                          trend='nc',
                                          fit_kw={'method': 'css-mle',
                                                  'maxiter': 5000,
                                                  'solver': 'lbfgs'})

print(auto_select.aic_min_order)

sns.heatmap(auto_select['aic'], annot=True, fmt='.1f')
plt.ylim([auto_select['aic'].shape[0], 0])
plt.title('AIC auto select')
plt.show()
```

```python
model = ARMA(df['ge'].values, order=(4, 2)).fit(trend='nc',
                                          maxiter=10000,
                                          start_params=np.array([0, 0, 0, 0, 0, 0]))
model.summary2()
```

```python
model.sigma2
```

```python
from sklearn.model_selection import TimeSeriesSplit

ts = TimeSeriesSplit(n_splits=3)
splits = [split for split in ts.split(df['ge'].values)]

best_order = (0, 0)
best_mse = 10**10
scores = []

for ar_order in range(5):
    for ma_order in range(5):
        try:
            mses = []
            for i in range(ts.n_splits):
                y_train = df['ge'].values[splits[i][0]]
                y_test = df['ge'].values[splits[i][1]]

                model_cv = ARMA(y_train, order=(ar_order, ma_order))
            
                result_cv = model_cv.fit(start_params=np.zeros(model_cv.k_ar+model_cv.k_ma),
                               solver='lbfgs', method='css', trend='nc', ic='BIC')
                predictions_test = result_cv.predict(start=len(y_train), end=len(y_train)+len(y_test)-1)
                mses.append(((y_test - (predictions_test))**2).mean())
                
            mse_mean = np.mean(mses)
            scores.append((model_cv.k_ar, model_cv.k_ma, mse_mean))
            print(model_cv.k_ar, model_cv.k_ma, mse_mean)
            if mse_mean < best_mse:
                best_mse = mse_mean
                best_order = (model_cv.k_ar, model_cv.k_ma)

        except:
                scores.append((model_cv.k_ar, model_cv.k_ma, np.nan))
                print(model_cv.k_ar, model_cv.k_ma, np.nan)
                
print(best_order, best_mse)

df_r = pd.DataFrame(scores, columns=['ar', 'ma', 'mse_cv'])
df_r.groupby(['ar', 'ma'])[['mse_cv']].max().sort_values(by='mse_cv').plot(kind='barh', figsize=(8, 8))
plt.xscale('log')
plt.show()
```

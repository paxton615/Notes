### Benefits of centering data

The primary benefit of centering your predictor data in linear modeling is that **the intercept represents the estimate of the target when all predictors are at their mean value.**

```python
data = pd.read_csv('/Users/paxton615/GA/resource-datasets/football_combine/combine_train.csv')
```

```python
df = data[['Name','Position','HeightInchesTotal','Weight']]
```

```python
print(df.dtypes)
print(df.shape)
print(df.isnull().sum())
df.describe()
```

```python
df['height_cm'] = df['HeightInchesTotal'].map(lambda x:x*2.54)
```

```python
df['weight_kg'] = df['Weight'].map(lambda x:x*0.453)
```

```python
df[['height_cm','weight_kg']].hist(figsize=(8,4));
```

```python
df[['height_cm','weight_kg']].plot(kind='box', figsize=(12,6));
```

- 在box plot中就会出现一个问题。如果量级差别太大，则某些图形显示时就是畸形的。比如这里的height_cm, 明显矮了很多。

```python
y,X = patsy.dmatrices('weight_kg ~ height_cm -1', data=df, return_type='dataframe')

lr = LinearRegression()
lr.fit(X, y)
print('intercept:', lr.intercept_)
print("coef", lr.coef_)
```

The intercept, -303.3, is the estimated weight of a baseball player when their height is 0 cm. The height coefficient 2.2044 is how much higher a player's weight in kg can be expected to be for each cm increase in height.

```python
df['height_ctr'] = df.height_cm - df.height_cm.mean()
```

```python
y,X = patsy.dmatrices('weight_kg ~ height_ctr', data=df, return_type='dataframe')

lr = LinearRegression()
lr.fit(X, y)
print('intercept:', lr.intercept_)
print("coef", lr.coef_)
```

> The intercept, 111.228, is the estimated weight of a baseball player of AVERAGE height (188 cm). The height coefficient 2.2 is how much higher a player's weight in pounds can be expected to be for each cm increase in height. The centering does not have any effect on the coefficient.

### 标准化的好处
#### 1. 截距就是被预测值的均值（如上例）。
2. 各特征按自己std的倍数缩放，一目了然，方便比较。
3. 有助于损失函数的优化和速度。
4. 很多模型的前期准备工作都要包含标准化：比如正则化之前要标准化特征；K-nearest中标准化让各特征贡献平衡；K-means也是一样；逻辑回归，神经网络，支持向量机也是同样原因。
5. 标准化只缩放纲量，而不改变数据分布。

```python
sns.jointplot(df.height_cm, df.weight_kg);
```

```python
sns.jointplot(df.height_cm-df.height_cm.mean(),\
              df.weight_kg-df.weight_kg.mean());
```

- 注意看标准化之后，x,y坐标轴的缩放变化。

#### Sklearn StandardScaler
---

The most common method of scaling is standardization. In standardization we first center the data, then we divide by the standard deviation to enforce that the standard deviation of the variable is one:

$$ X_{std} = \frac{X - \bar{X}}{\sigma_{X}} $$

```python
scaler = StandardScaler()

Xstd = scaler.fit_transform(df[['height_cm', 'weight_kg']])
```

```python
print(Xstd.mean(axis=0))
print(Xstd.std(axis=0))
```

```python
df[['height_cm', 'weight_kg']].describe()
```

```python
pd.DataFrame(Xstd, columns=['height_std','weight_std']).describe()
```

## Min-Max Scaling

---

This scaling limits a variable's value to lie between 0 and 1. Think of this as squishing the variable to be constrained to a specific range.

$$ X_{\rm minmax} = \frac{X - {\rm min}(X)}{{\rm max}(X) - {\rm min}(X)} $$


Typically standardization is preferred to min-max scaling. However, it suffers in the case of extremely low standard deviations and min-max scaling might be a good choice.

```python
mms = MinMaxScaler()
Xmms = mms.fit_transform(df[['height_cm','weight_kg']])
Xmms = pd.DataFrame(Xmms, columns = ['height_cm','weight_kg'])
```

```python
Xmms.describe()
```

```python
print(Xmms.std(axis=0))
```

```python
sns.jointplot(Xmms.height_cm, Xmms.weight_kg);
```

- 关注点依然是坐标轴的量级变化

```python
wine = pd.read_csv('/Users/paxton615/GA/resource-datasets/wine_data/winequality_merged.csv')
```

```python
wine.columns
```

```python
quality = wine.quality
quality.describe()
```

```python
X = wine[['fixed acidity','residual sugar','sulphates','alcohol']]
X.head()
```

##### 使用scaler, mms, 和 PolymialFeatures（在原特征基础上构造0-n次方的特征，即构造原特征的高次方特征）

```python
Xs = scaler.fit_transform(X)
Xm = mms.fit_transform(X)
```

```python
#Don't include bias in the polynomial features since Linear Regression will take care of that.
pf = PolynomialFeatures(degree=2, include_bias=False) #在原特征基础上创造0-2次方对应的原特征出来
Xpf = pf.fit_transform(Xs) 
# 注意这里实在 Xs 的基础上构建多次方特征！！  然后再标准化！！
scaler_p = StandardScaler()
Xpfs = scaler_p.fit_transform(pf.fit_transform(Xs))
```

```python
print(X.shape, Xs.shape, Xm.shape, Xpf.shape, Xpfs.shape)
```

```python
Xpfs
```

```python
Xpf
```

```python
lr = LinearRegression()

raw_r2 = cross_val_score(lr, X, quality, cv=10)
std_r2 = cross_val_score(lr, Xs,quality, cv=10)
mms_r2 = cross_val_score(lr, Xm,quality, cv=10)
pf_r2 = cross_val_score(lr, Xpf,quality, cv=10)
pfs_r2 = cross_val_score(lr, Xpfs,quality, cv=10)
```

```python
print('Original mean CV-R2: {:.3}, std CV-R2: {:.3}'.format(np.mean(raw_r2),np.std(raw_r2)))
print('Standardized mean CV-R2: {:.3}, std CV-R2: {:.3}'.format(np.mean(std_r2),np.std(std_r2)))
print('Min-max mean CV-R2: {:.3}, std CV-R2: {:.3}'.format(np.mean(mms_r2), np.std(mms_r2)))
print('Polynomial features mean CV-R2: {:.3}, std CV-R2: {:.3}'.format(np.mean(pf_r2), np.std(pf_r2)))
print('Standardized Polynomial features mean CV-R2: {:.3}, std CV-R2: {:.3}'.format(np.mean(pfs_r2), np.std(pfs_r2)))
```

```python
Mraw = lr.fit(X, quality)
Mstd = lr.fit(Xs, quality)
Mmms = lr.fit(Xm, quality)
Mpf = lr.fit(Xpf, quality)
Mpfs = lr.fit(Xpfs, quality)
print("Mraw:",Mraw.intercept_, Mraw.coef_)
print('-'*50)
print("Mstd:",Mstd.intercept_, Mstd.coef_)
print('-'*50)
print("Mmms:",Mmms.intercept_, Mmms.coef_)
print('-'*50)
print("Mraw:",Mpf.intercept_, Mpf.coef_)
print('-'*50)
print("Mraw:",Mpfs.intercept_, Mpfs.coef_)
```
